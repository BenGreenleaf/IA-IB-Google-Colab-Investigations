```python
import numpy as np
import control as ctr

H0kp = 102.011
Td = 2
m=200

G = ctr.tf([Td*H0kp,H0kp], [m,m,0,0])
print(ctr.margin(G))
out = ctr.nyquist_plot(G)
out = ctr.bode_plot(G)
```
