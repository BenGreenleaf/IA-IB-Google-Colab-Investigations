```python
import control as ctr
import numpy as np
import matplotlib.pyplot as plt
```

```python
G1 = ctr.tf([22000], [1,22000])
out = ctr.bode_plot(G1)
```

```python
G2 = ctr.tf([22000,0], np.convolve([1,300],[1,22000]))
out = ctr.bode_plot(G2)
```

```python
t1, y1 = ctr.step_response(G1)
fig3 = plt.figure()
plt.plot(t1, y1)
plt.grid()
```

```python
t2, y2 = ctr.step_response(G2)
fig4 = plt.figure()
plt.plot(t2, y2)
plt.grid()
```
